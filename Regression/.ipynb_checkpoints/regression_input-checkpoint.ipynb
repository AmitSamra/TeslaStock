```python
import os
import numpy as np
import pandas as pd
import statsmodels.formula.api as sm
```

### Linear Regression

```python
df_merged = pd.read_csv('/Users/amit/Coding/Projects/TeslaStock/Regression/final_df.csv', parse_dates=['date'])
df_merged
```

```python
df_merged = df_merged.filter(['date', 'percent_change', 'tweet_count', 'number_articles', 'daily_sentiment_score'], axis=1)
df_merged = df_merged.drop(0, axis=0)
df_merged = df_merged.set_index('date')
df_merged
```

```python
df_merged.corr()
```

```python
# Linear Regression Model
result = sm.ols(formula="percent_change ~ number_articles + daily_sentiment_score + tweet_count", data=df_merged).fit()
```

```python
print(result.summary())
```

### Linear Regression Using Absolute Value

```python
# Use abs value of percent_change and daily_sentiment_score
df_merged2 = df_merged.copy()
df_merged2['percent_change'] = abs(df_merged2['percent_change'])
df_merged2['daily_sentiment_score'] = abs(df_merged2['daily_sentiment_score'])
df_merged2
```

```python
df_merged2.corr()
```

```python
# Linear Regression Model Using Absolute Value
result_abs = sm.ols(formula="percent_change ~ number_articles + daily_sentiment_score + tweet_count", data=df_merged2).fit()
```

```python
print(result_abs.summary())
```
